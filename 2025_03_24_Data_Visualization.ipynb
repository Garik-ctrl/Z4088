### First drawing

In order to **<font color='##13e5e1'>create a plot</font>**, we commonly utilize the function `plt.subplots()`, which returns two objects:
* `fig` - serves as a container for the plots
* `ax/axes` - represents the actual plot/plots themselves

To **<font color='##13e5e1'>incorporate visual elements</font>** onto the plot, we utilize various operations on the `ax` object, e.g. by calling methods such as `.plot()`, we can add different types of visualizations, such as line plots.

Once the desired visualizations have been added, we can display the final graph by using the `plt.show()` function. It **<font color='##13e5e1'>displays the graph</font>** in a separate window or within the notebook interface.

````python
# data to vizualize
x = [0, 1, 2, 3]
y = [0, 1, 2, 3]

# create a canva (plot container)
fig, ax = plt.subplots()
ax.plot(x, y) # line plot
ax.grid()
plt.show() # display the graph
````

To add a grid to a matplotlib plot, you can use the `.grid()` method. Such grid is a visual element that can be added to a plot to enhance its appearance and aid in data interpretation. It consists of **<font color='##13e5e1'>horizontal and vertical lines</font>** that align with the major ticks on the X-axis and Y-axis.

By default, the `.grid()` function enables the gridlines with a light gray color, but you have the flexibility to **<font color='##13e5e1'>customize the grid</font>** by adjusting its appearance. The `.grid()` function accepts optional parameters such as
* `color`
* `linestyle`
* `linewidth`
* `alpha`

to modify the color, line style, thickness, and transparency of the gridlines.

````python
fig, ax = plt.subplots()

ax.grid()
plt.show()
````

````python
fig,ax = plt.subplots()

ax.grid(
    color='darkgreen',
    linestyle='--',
    #linewidth=5,
    alpha=0.9
)

plt.show()
````

There are many additonal paramters, see official documentation for more: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html

For example, the `axis` parameter in the `.grid()` method allows you to control which axis the gridlines are drawn on:
* `axis='x'` to add the gridlines only along the X-axis
* `axis='y'` to add them only along the Y-axis.

To include gridlines on both axes, you can set `axis='both'` which is the default.

````python
fig,ax = plt.subplots()

ax.grid(
    axis='x',
    color='#7bf992',
    linewidth=1.5,
)

ax.grid(
    axis='y',
    color='b',
    linewidth=1.5,
)

plt.show()
````

## Weather in Seattle

The best way to show the utility of this library is by using real data.

*Instead of using the URL below, you can also download the file from: https://drive.google.com/drive/folders/1Tx2SceHPlAu6qy_ox1j1kcfw8uhELWDU?usp=drive_link*

````python
url='https://raw.githubusercontent.com/matzim95/ML-datasets/master/weather.csv'

weather=pd.read_csv(
    url,
    parse_dates=True,
    index_col='date'
)
````

````python
weather.head()
````

### Basic time series vizualization

````python
fig, ax = plt.subplots()

# draw a plot using an "axis" object
ax.plot(weather.index, weather['temp_max'])
plt.show()
````

### Adjust size of the canva
Note that the labels on the chart are not readable and the default size of the chart is not optimal.

````python
fig, ax = plt.subplots(figsize=(15, 2))

ax.plot(weather.index, weather['temp_max'])
plt.show()
````

### Display multiple plots/layers on the same graph

What if you want to draw **<font color='##13e5e1'>two or more plots on a single graph</font>** in matplotlib, allowing for visual comparison and analysis of multiple datasets or features within the same figure?

````python
fig, ax=plt.subplots(figsize=(12, 4))

ax.plot(weather.index, weather['temp_max'], color='#7bf992')
ax.plot(weather.index, weather['temp_min'])
ax.plot(weather.index, weather['precipitation'])

plt.show()
````

### Specify legend

Note that we can only **<font color='##13e5e1'>guess which color corresponds to which column</font>**. Here it's easy because the data comes from one region (ie. Seattle) and describes the minimum and maximum temperature. What if they came from different places or described traits that would not be so obvious to distinguish based on the graph?

If you want to **<font color='##13e5e1'>add legend labels to a graph</font>** in matplotlib, allowing you to distinguish and identify different datasets or plot elements in the visualization, follow these steps:
* assign a label to each plot
* after plotting all the necessary elements, call the `plt.legend()` function to display the legend on the graph.

Customize the appearance of the legend, if desired, by using optional parameters available in the `plt.legend()` function. These parameters allow you to control the position, alignment, font size, and other properties of the legend.

Display the graph using the `plt.show()` function.

````python
fig, ax=plt.subplots(figsize=(12, 4))

ax.plot(weather.index, weather['temp_max'], label='max')
ax.plot(weather.index, weather['temp_min'], label='min')

plt.legend()
plt.show()
````

### Markers and lines on the graph

In example below, let's consider only partial data for 20 consecutive temperature values. Note, that rotaion of the X-axis ticks can also help to make them readable.

````python
n = 20

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(
    weather.index[:n],
    weather['temp_max'][:n],
)

ax.plot(
    weather.index[:n],
    weather['temp_min'][:n],
)

plt.xticks(rotation=45, ha='right')
plt.show()
````

So far we've been using the default arguments but you can have a huge impact on what the chart you create looks like!

In matplotlib, **<font color='##13e5e1'>markers and lines are used to represent data points and connect them</font>** in a plot:
* **<font color='darkorange'>markers</font>** are symbols or shapes that represent individual data points on a plot
 * they can be used to highlight the position of data values
 * there is a variety of marker options, such as dots, squares, triangles, and more
 * customization options for markers include size, color, edge color, and transparency
* **<font color='darkorange'>lines</font>** are used to connect data points in a plot, forming continuous lines or curves - they help to visualize trends and relationships between data points
 * common line styles include solid lines, dashed lines, dotted lines, and more
 * you can customize line colors, widths, and transparency using parameters like color, linewidth, and alpha

````python
n = 20

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(
    weather.index[:n],
    weather['temp_max'][:n],
    marker='o',
    linestyle='--',
    color='r'
)

ax.plot(
    weather.index[:n],
    weather['temp_min'][:n],
    marker='v',
    linestyle='None',
    color='b'
)

plt.xticks(rotation=45)
plt.show()
````

### Subplots

Subplots in matplotlib allow you to **<font color='##13e5e1'>create multiple plots within a single figure</font>**. Using the `plt.subplots()` function, you can specify the layout of subplots, access each subplot individually, plot data on each subplot, and customize them. Subplots are useful for comparing datasets or displaying different aspects of the same data.

````python
n = 50
fig, ax = plt.subplots(2, 1,sharex=True, figsize=(10, 5))

ax[0].plot(austin_weather.index[:n], austin_weather['TempAvgF'][:n], color='r')
ax[0].plot(austin_weather.index[:n], austin_weather['TempHighF'][:n], color='r', linestyle='--')
ax[0].plot(austin_weather.index[:n], austin_weather['TempLowF'][:n], color='r', linestyle='dotted')


ax[1].plot(austin_weather.index[:n], austin_weather['DewPointAvgF'][:n].astype(float), color='b')
ax[1].plot(austin_weather.index[:n], austin_weather['DewPointHighF'][:n].astype(float), color='b', linestyle='--')
ax[1].plot(austin_weather.index[:n], austin_weather['DewPointLowF'][:n].astype(float), color='b', linestyle='dotted')

ax[1].set_xlabel('Date (year-month-day)')
ax[0].set_ylabel('Temperature')
ax[1].set_ylabel('Dew Point')

ax[0].set_title('Weather in Austin')
ax[1].xaxis.set_tick_params(rotation=45)
plt.show()
````

### Two distinct scales on Y-axis

````python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(
    merged.index,
    merged.Interpolated,color='b'
)
ax.set_xlabel('Time')
ax.set_ylabel('$CO_2$', color='b')
ax.tick_params('y', color='b', labelcolor='b')

# clone Y-axis (leave X-axis as shared one)
ax2 = ax.twinx()
ax2.plot(
    merged.index,
    merged.AverageTemperature.rolling(window=365, center=True,min_periods=0).mean(),
    color='r'
)
ax2.set_ylabel('Temperature', color='r')
ax2.tick_params('y', color='r', labelcolor='r')

plt.show()
````

### Create charts using pre-defined custom functions

````python
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
  axis.plot(x, y, color)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel, color=color)
  axis.tick_params('y', colors=color, labelcolor=color)
````

````python
fig,ax = plt.subplots()

plot_timeseries(
    axis=ax,
    x=merged.index,
    y=merged.Trend,
    color='r',
    xlabel='Time',
    ylabel='CO2'
)

ax2=ax.twinx()
plot_timeseries(
    axis=ax2,
    x=merged.index,
    y=merged.AverageTemperature,
    color='b',
    xlabel='Time',
    ylabel='Temperature'
)
````

Bar charts are commonly used to display survey results, sales data, or any data with distinct categories. Matplotlib provides flexibility in customizing the appearance of bar charts, allowing you to create visually appealing and informative visualizations.

## Histograms

A histogram is a **<font color='##13e5e1'>graphical representation of the distribution of a dataset</font>**. It provides a visual summary of the underlying frequency or probability distribution of a set of continuous or discrete data.

Histograms are commonly used in data analysis and exploratory data visualization. They allow for a quick assessment of the distribution of data and can highlight important features, such as outliers or clustering tendencies. Histograms are especially useful for analyzing continuous numerical data, such as exam scores, population densities, or rainfall measurements.

````python
url = 'https://raw.githubusercontent.com/matzim95/ML-datasets/master/olympic.csv'

medals = pd.read_csv(url, index_col='Year')
medals.head()
````

````python
def get_dict_with_grouped_data(df, column):
  groupby = df.groupby(column)
  return {x: groupby.get_group(x) for x in groupby.groups}
````

## Error bars (in barplots & lineplots)

Error bars are used to **<font color='##13e5e1'>represent the uncertainty or variability</font>** associated with the data points. They provide a visual indication of the confidence intervals, standard errors, or standard deviations around the mean or median values.

Here's how to add error bars to a bar/line plot in matplotlib:


````python
fig, ax = plt.subplots()

ax.bar(
    'Gymnastics',
    all_sports['Gymnastics']['Height'].mean(),
    yerr=all_sports['Gymnastics']['Height'].std()*2
)
ax.bar(
    'Basketball',
    all_sports['Basketball']['Height'].mean(),
    yerr=all_sports['Basketball']['Height'].std()*2
)

ax.set_ylabel('Height [cm]')
plt.show()
````

## Save a graph

To save a graph created using matplotlib, you can use the `plt.savefig()` function. Here's how you can do it:
* after creating your graph, add the following line of code before calling `plt.show()` or at the end of your script
```python
plt.savefig('filename.png')
```
Replace `'filename.png'` with the desired name, file extension and path to the file. You can specify various file formats such as PNG, JPEG, PDF, SVG, etc.

Make sure to include the appropriate path if you want to save the graph in a specific directory. For example, `'path/to/save/directory/filename.png'`.




````python
fig.set_size_inches([10,6])
fig.savefig('temp_co2_60_10.png')
````

The `plt.savefig()` function provides additional parameters that allow you to customize the saved figure. Here are some commonly used parameters:
* `dpi` (dots per inch): Specifies the resolution or quality of the saved figure. Higher values result in a higher-resolution image
* `bbox_inches`: Controls the portion of the figure to save. By default, it saves the entire figure, but you can specify different options like 'tight' to remove the extra whitespace around the plot. Example: plt.savefig('filename.png', bbox_inches='tight')
* `transparent`: Determines if the saved figure has a transparent background.
* `pad_inches`: Adds padding around the saved figure to prevent elements from being cut off. You can specify the amount of padding in inches. Example: plt.savefig('filename.png', pad_inches=0.5)

## Histogram with KDE

In Seaborn, `displot()` is a function used to create a variety of plots for univariate distributions. It supports various types of plots, including histograms, **<font color='##13e5e1'>kernel density estimation (KDE) plots</font>**, and rug plots. When using `displot()` with the kind parameter set to 'kde', it generates a KDE plot.

A KDE plot visualizes the **<font color='##13e5e1'>estimated probability density function of a continuous variable</font>**. It provides a smooth curve that approximates the underlying distribution of the data.

````python
sns.displot(wine.Alcohol, kde=True)
plt.show()
````
